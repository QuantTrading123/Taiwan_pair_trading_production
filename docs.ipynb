```python
import sys
import random
from PyQt5.QtGui import QPainter, QColor
from PyQt5.QtCore import Qt, QTimer, QTime
from PyQt5.QtWidgets import QWidget, QApplication

class Heatmap(QWidget):

    def __init__(self):
        super().__init__()

        self.bids = []
        self.asks = []
        self.data_length = 5
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_data)
        self.timer.start(1000) # 每秒更新一次數據

    def update_data(self):
        self.bids = [random.randint(1, 10) for _ in range(self.data_length)]
        self.asks = [random.randint(1, 10) for _ in range(self.data_length)]
        self.update()

    def paintEvent(self, event):
        qp = QPainter()
        qp.begin(self)
        self.draw_heatmap(qp)
        qp.end()

    def draw_heatmap(self, qp):
        w = self.width()
        h = self.height()

        # 繪製背景
        qp.fillRect(0, 0, w, h, QColor(255, 255, 255))

        # 設置熱度圖的顏色
        heatmap_colors = [
            QColor(255, 255, 255),
            QColor(255, 204, 204),
            QColor(255, 153, 153),
            QColor(255, 102, 102),
            QColor(255, 51, 51),
            QColor(255, 0, 0)
        ]

        # 繪製熱度圖
        for i, bid in enumerate(self.bids):
            x = (i + 1) * w // (self.data_length + 1)
            for j in range(bid):
                y = (j + 1) * h // (max(self.bids + self.asks) + 1)
                color = heatmap_colors[min(j, len(heatmap_colors) - 1)]
                qp.fillRect(x - 10, h - y - 10, 20, 20, color)

        for i, ask in enumerate(self.asks):
            x = (i + 1) * w // (self.data_length + 1)
            for j in range(ask):
                y = (j + 1) * h // (max(self.bids + self.asks) + 1)
                color = heatmap_colors[min(j, len(heatmap_colors) - 1)]
                qp.fillRect(x - 10, h - y - 10, 20, 20, color)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    heatmap = Heatmap()
    heatmap.show()
    sys.exit(app.exec_())
```
